## Outlier Treatment

#### What is an outlier: 
it's a data point that's very far from the other data points. Mathematically, an outlier is usually defined as an observation more than three standard deviations from the mean

-- Outlier can present in the dataset for both error entry as well as for an actual exceptional true value. However, presence of an outlier in a dataset can lead to a non-performing statistical analysis as it generates bias in the dataset

--Outlier normally can be detected by various techiques such as
                          
    a. Scatter Plot
    b. Box Plot
    c. Z score
    d. IQR or interquartile range

### Boxplot

```python
sns.boxplot(x=df['GrLivArea'])
```

From the above boxplot we can see that there are lot of outliers are present in the column

### Scatterplot

```python
sns.scatterplot(x='SalePrice', y='GrLivArea', data=df)
```

### Detecting outlier using Z score

Formula for Z score = (Observation  Mean)/Standard Deviation

z = (X  ) / 

```python
# Since we are discussing the outlier by taking an example of column 'GrLivArea', let us remove outlier by using z-score from the
# dataset for this particular column
from scipy import stats
#df1 = df[df_num]
df1 = df[(np.abs(stats.zscore(df.GrLivArea)) < 3)]
```

Once, we remove the outlier, let us see the scatterplot once again

```python
sns.scatterplot(x='SalePrice', y='GrLivArea', data=df1)
```

### InterQuantile Range (IQR)

```python
Q1 = df.GrLivArea.quantile(0.25)
Q3 = df.GrLivArea.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df1 = df[~((df['GrLivArea'] < (Q1 - 1.5 * IQR)) |(df['GrLivArea'] > (Q3 + 1.5 * IQR)))]
```

```python
sns.scatterplot(x='SalePrice', y='GrLivArea', data=df1)
```

We can see that outlier is removed from 'GrLivArea' column
